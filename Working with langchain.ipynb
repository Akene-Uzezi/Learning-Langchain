```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
```

```python
import os
model = os.environ.get("modelName")
apiKey = os.environ.get("OPENROUTER_API_KEY")
chat = ChatOpenAI(model=model, model_kwargs={ 'seed': 365 },
    temperature=0,
    base_url='https://openrouter.ai/api/v1',
    api_key=apiKey,
    max_completion_tokens=100
)
```

```python
response = chat.invoke('I recently adopted a dog. suggest some names')
```

```python
print(response.content)
```
