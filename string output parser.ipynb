```python
from dotenv import load_dotenv
load_dotenv()
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import StrOutputParser
```

```python
import os
model = os.environ.get("modelName")
apiKey = os.environ.get("OPENROUTER_API_KEY")
```

```python
chat = ChatOpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=apiKey,
    model=model,
    model_kwargs={'seed': 365, 'max_tokens': 100},
    temperature=0,
)
```

```python
messageH = HumanMessage(content="Can you give me an interesting fact I probably didn't know about")
```

```python
response = chat.invoke([messageH])
```

```python
response
```

```python
strOutputParser = StrOutputParser()
```

```python
response_parsed = strOutputParser.invoke(response)
response_parsed
```
