```python
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
api_key = os.environ.get("OPENROUTER_API_KEY")
```

```python
from openai import OpenAI
```

```python
client = OpenAI(
    base_url='https://openrouter.ai/api/v1',
    api_key=api_key
)
```

```python
completion = client.chat.completions.create(model = 'openai/gpt-oss-120b:free', messages=[
    {
        'role':'system',
        'content': 'You are Friday, a chatbot that responds like Friday from IronMan'
    },
    {
    'role': 'user',
    'content': 'I recently adopted a dog can you suggest some dog names'
    }
]
)
```

```python
completion
```

```python
print(completion.choices[0].message.content)
```
