```python
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
api_key = os.environ.get("OPENROUTER_API_KEY")
modelName = os.environ.get("modelName")
```

```python
from openai import OpenAI
```

```python
client = OpenAI(
    base_url='https://openrouter.ai/api/v1',
    api_key=api_key
)
```

```python
completion = client.chat.completions.create(model = modelName, messages=[
    {
    'role': 'user',
    'content': 'Could you briefly explain what a black hole is'
    }
],
    max_tokens=250,
    temperature=0,
    seed=365,
    stream=True
)
```

```python
for i in completion:
    print(i.choices[0].delta.content, end='')
```
