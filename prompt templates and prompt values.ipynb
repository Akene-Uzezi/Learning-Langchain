```python
TEMPLATE = '''
System:
{description}

Human:
I've recently adopted a {pet}.
Could you suggest some {pet} names?
'''
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
promptTemplate = PromptTemplate.from_template(template = TEMPLATE)
```

```python
promptTemplate
```

```python
promptValue = promptTemplate.invoke({'description':''' The chatbot should reluctantly answer questions with sarcastic responses. ''',
    'pet': 'dog'}
)
```

```python
promptValue
```
