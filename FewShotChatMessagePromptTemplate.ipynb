```python
from dotenv import load_dotenv
import os
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import (ChatPromptTemplate,HumanMessagePromptTemplate,AIMessagePromptTemplate,FewShotChatMessagePromptTemplate)
```

```python
load_dotenv()
```

```python
apiKey = os.environ.get("OPENROUTER_API_KEY")
```

```python
model = os.environ.get("modelName")
```

```python
chat = ChatOpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=apiKey,
    model_name = model,
    temperature = 0,
    max_completion_tokens=100,
    model_kwargs={'seed': 365}
)
```

```python
TEMPLATEH = '''I recently adopted a {pet}. Could you suggest some {pet} names '''
```

```python
TEMPLATEAI = '''{response}'''
messageh = HumanMessagePromptTemplate.from_template(template=TEMPLATEH)
messageai = AIMessagePromptTemplate.from_template(template=TEMPLATEAI)
```

```python
exampleTemplate = ChatPromptTemplate.from_messages([messageh, messageai])
```

```python
examples = [
    {
        'pet': 'dog',
        'response': 'Of course why the fuck not? 🙄'
    },
    {
        'pet': 'cat',
        'response': 'Of course why the fuck not? 🙄'
    }
]
```

```python
fewShotPrompt = FewShotChatMessagePromptTemplate(examples=examples, example_prompt=exampleTemplate, input_variables=['pet'])
```

```python
chatTemplate = ChatPromptTemplate.from_messages([fewShotPrompt, messageh])
```

```python
chatValue = chatTemplate.invoke({'pet': 'rabbit'})
```

```python
chatValue
```

```python
for i in chatValue.messages:
    print(f'{i.type}: {i.content}\n')
```

```python
response = chat.invoke(chatValue)
```
