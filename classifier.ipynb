```python
from dotenv import load_dotenv
import os
from openai import OpenAI
```

```python
load_dotenv()
```

```python
apiKey = os.environ.get("OPENROUTER_API_KEY")
```

```python
client = OpenAI(
    base_url='https://openrouter.ai/api/v1',
    api_key=apiKey
)
```

```python
completion = client.chat.completions.create(model='openai/gpt-oss-120b:free', 
    messages=[
        {
            'role': 'system',
            'content': 'You are a sentiment Analysis Assistant. Classify text to Positive, Neutral or Negative'
        },
        {
            'role': 'user',
            'content': 'I love this product'
        },
        {
            'role': 'assistant',
            'content': 'Positive'
        },
        {
            'role': 'user',
            'content': 'It arrived yesterday'
        },
        {
            'role': 'assistant',
            'content': 'Neutral'
        },
        {
            'role': 'user',
            'content': 'The product is garbage'
        },
        {
            'role': 'assistant',
            'content': 'Negative'
        },
        {
            'role': 'user',
            'content': 'I need more of this'
        }
    ]
)
```

```python
print(completion.choices[0].message.content)
```
