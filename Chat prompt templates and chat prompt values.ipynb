```python
from dotenv import load_dotenv
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import (SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate)
import os
```

```python
load_dotenv()
```

```python
apiKey = os.environ.get("OPENROUTER_API_KEY")
model = os.environ.get("modelName")
```

```python
chat = ChatOpenAI(
    model=model,
    base_url = 'https://openrouter.ai/api/v1',
    api_key=apiKey,
    model_kwargs={'seed': 365},
    temperature=0,
    max_completion_tokens=100
)
```

```python
TEMPLATES = '{description}'
TEMPLATEH = ''' Ive recently adopted a {pet}
Could you suggest some {pet} names '''

messageTemplateS = SystemMessagePromptTemplate.from_template(template=TEMPLATES)
messageTemplateH = HumanMessagePromptTemplate.from_template(template=TEMPLATEH)
```

```python
messageTemplateH
```

```python
chatTemplate = ChatPromptTemplate.from_messages([messageTemplateS, messageTemplateH])
```

```python
chatTemplate
```

```python
chatValue = chatTemplate.invoke({
    'description': '''The chatbot should reluctantly answer questions with sarcastic responses ''',
    'pet': '''dog'''
})
```

```python
chatValue
```

```python
response = chat.invoke(chatValue)
```
